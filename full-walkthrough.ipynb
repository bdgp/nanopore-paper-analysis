# 1: System dependencies

If you are running Ubuntu 14.04 LTS here are the required dependencies:

    sudo apt-get update
    sudo apt-get install python-virtualenv python-dev zlib1g-dev libhdf5-dev r-base-dev git parallel default-jre
    sudo apt-get install bioperl --fix-missing

If you are running CentOS6/RHEL6:

    sudo yum install python-virtualenv screen zlib-devel hdf5-devel R readline-devel perl-CPAN git parallel
    sudo cpan install Statistics::Descriptive
    sudo cpan install Module::Build

# 2: Prepare environment

Download the example Makefile:

    wget https://raw.githubusercontent.com/jts/nanopore-paper-analysis/master/full-pipeline.make

Set up a new Python virtualenv (the makefile will install various python packages)

    virtualenv venv
    source venv/bin/activate


# 3: Run the pipeline

The assembly makefile pipeline will:

* download the E. coli nanopore read sets from the ENA
* install further Python and C dependencies
* correct the reads in parallel using nanocorrect
* assemble the genome with Celera assembler
* polish the assembly with nanopolish

You will need to set CORES to a value sensible for your system (suggest # CPUs - 1)

    make -f full-pipeline.make polished_genome.fasta CORES=90

This process takes about 24 hours to run to completeness on a 90-core 3Ghz server.
